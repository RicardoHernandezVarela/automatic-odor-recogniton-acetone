```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

acetone_1 = pd.read_csv('muestras1ml.csv', delimiter=",",  header=None)
acetone_2 = pd.read_csv('muestras2ml.csv', delimiter=",",  header=None)
acetone_3 = pd.read_csv('muestras02ml.csv', delimiter=",",  header=None)
acetone_4 = pd.read_csv('muestras04ml.csv', delimiter=",",  header=None)
```

```python
acetone_1 = acetone_1.drop(acetone_1.index[0]) #eliminar fila
acetone_2 = acetone_2.drop(acetone_2.index[0])
acetone_3 = acetone_3.drop(acetone_3.index[0])
acetone_4 = acetone_4.drop(acetone_4.index[0])
```

```python
acetone_4.head()
```

```python
acetone_3.head()
```

```python
frames = [acetone_1, acetone_2, acetone_3, acetone_4]
```

```python
samples = pd.concat(frames)
samples.head()
```

```python
samples = samples.drop([0], axis=1) #eliminar columna
samples.head()
```

```python
samples['Clase'] = samples[385]
samples.head()
```

```python
samples = samples.drop([385], axis=1)
samples
```

```python
#samples = samples.sample(frac=1).reset_index(drop=True)
samples = samples.reset_index(drop=True)
samples.to_csv('acetone_samples.csv', index=False)
```

```python
samples.head()
```

```python
clases = samples['Clase']
samples = samples.drop(['Clase'], axis=1)
samples.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, make_scorer, accuracy_score

originalclass = []
predictedclass = []


def classification_report_cv(y_true, y_pred):
    originalclass.extend(y_true)
    predictedclass.extend(y_pred)
    print(classification_report(originalclass, predictedclass)) 
    return accuracy_score(y_true, y_pred)

def confusion_matrix_cv(y_true, y_pred):
    originalclass.extend(y_true)
    predictedclass.extend(y_pred)
    print(confusion_matrix(originalclass, predictedclass)) 
    return accuracy_score(y_true, y_pred)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(samples, clases, test_size=0.4, random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=1, probability=True, gamma='auto'),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis()]
```

```python
from sklearn.model_selection import cross_val_score
```

```python
knn = cross_val_score(classifiers[0], samples, clases, cv=10)
knn.mean()
```

```python
svm = cross_val_score(classifiers[1], samples, clases, cv=10)
svm.mean()
```

```python
nb = cross_val_score(classifiers[6], samples, clases, cv=10)
nb.mean()
```

```python
rf = cross_val_score(classifiers[3], samples, clases, cv=10)
rf.mean()
```

```python
#Importar PCA y crear la instancia del modelo
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

#Aplicar PCA a los datos
componentes_principales = pca.fit_transform(samples)

#Crear un DataFrame con los valores de salida del PCA aplicado a los datos de los sensores
pca_data = pd.DataFrame(data = componentes_principales,
              columns = ['Componente principal 1', 'Componente principal 2'])

#Unir los datos de los sensores después de aplicar PCA y los valores etiquetados con las clases
final_data = pd.concat([pca_data, clases], axis = 1)
```

```python
final_data
```

```python
varianza_componentes = pca.explained_variance_ratio_
print(varianza_componentes)
varianza_total = sum(varianza_componentes)*100
print('La varianza total de los dos componentes es {}%'.format(varianza_total))
```

```python
#visualizar resultado de PCA
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Componente principal 1, var = {}%'.format(str(varianza_componentes[0]*100)), fontsize = 15)
ax.set_ylabel('Componente principal 2, var = {}%'.format(str(varianza_componentes[1]*100)), fontsize = 15)
ax.set_title('PCA 2 componentes', fontsize = 20)

clases = ['1', '2', '3', '4']

colores = ['#DDFF33', '#1c2345', '#3352FF', '#2e294e']

marcas = ["p", "P", "*", "*"]

for clase, color, marca in zip(clases, colores, marcas):
  indices = final_data['Clase'] == clase
  ax.scatter(final_data.loc[indices, 'Componente principal 1'], final_data.loc[indices, 'Componente principal 2'], c = color, s = 60, marker = marca)

ax.legend(clases)
ax.grid()

plt.show()
fig.savefig("acetonePCA.png")
```
